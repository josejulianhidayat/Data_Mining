```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

path='https://raw.githubusercontent.com/asepmuhidin/MK-Data-Mining-UPB/main/dataset/titanic.csv'

df=pd.read_csv(path)
df.head()
```

```python
df.duplicated().sum()
```

```python
df.drop('PassengerId', axis=1, inplace=True)
df.head()
```

```python
df.duplicated().sum()
```

```python
df.isna().sum()
```

```python
print(f"Jumlah Data Awal Adalah {df.shape[0]} Baris")
```

```python
df.drop(columns=['Cabin'], inplace=True)
```

```python
df.isna().sum()
```

```python
df.Age.mean()
```

```python
df.Age.median()
```

```python
df.Age.mode()
```

```python
df["Age"] = df["Age"].fillna(df["Age"].mean())
```

```python
df.Embarked.value_counts()
```

```python
df.Embarked.fillna('S', inplace=True)
```

```python
df.describe()
```

```python
jumlah_penumpang = df[df["Age"] > 29].shape[0]
```

```python
print("Jumlah penumpang umur di atas 29 tahun:", jumlah_penumpang)
```

```python
penumpang_tidak_bayar = df[df["Fare"] == 0]
```

```python
jumlah_penumpang_tidak_bayar = penumpang_tidak_bayar.shape[0]
```

```python
print("Jumlah penumpang yang tidak membayar tiket:", jumlah_penumpang_tidak_bayar)
```

```python
print(penumpang_tidak_bayar)
```

```python
# Mencari penumpang dengan umur terkecil
umur_terkecil = df['Age'].min()
penumpang_umur_terkecil = df[df['Age'] == umur_terkecil]

# Menampilkan hasil
print(penumpang_umur_terkecil[['Name', 'Age']])
```

```python
# Mencari penumpang dengan umur tertua
umur_tertua = df['Age'].max()
penumpang_umur_tertua = df[df['Age'] == umur_tertua]

# Menampilkan hasil
print(penumpang_umur_tertua[['Name', 'Age']])
```

```python
# Menampilkan semua penumpang yang berjenis kelamin perempuan
penumpang_perempuan = df[df['Sex'] == 'female']

# Menampilkan hasil
print(penumpang_perempuan)
```

```python
# Menampilkan semua penumpang yang berjenis kelamin perempuan
penumpang_perempuan = df[df['Sex'] == 'male']

# Menampilkan hasil
print(penumpang_perempuan)
```

```python
# Menampilkan penumpang yang tidak membayar (Fare = 0) dan selamat (Survived = 1)
penumpang_tidak_bayar_selamat = df[(df['Fare'] == 0) & (df['Survived'] == 1)]

# Menampilkan hasil
print(penumpang_tidak_bayar_selamat)
```

```python
# Menampilkan penumpang yang membayar (Fare > 0) dan tidak selamat (Survived = 0)
penumpang_bayar_tidak_selamat = df[(df['Fare'] > 0) & (df['Survived'] == 0)]

# Menampilkan hasil
print(penumpang_bayar_tidak_selamat)
```

```python
# Menampilkan penumpang yang membayar (Fare > 0) dan selamat (Survived = 1)
penumpang_bayar_dan_selamat = df[(df['Fare'] > 0) & (df['Survived'] == 1)]

# Menampilkan hasil
print(penumpang_bayar_dan_selamat)
```
