```python
#Mengaktifkan Library Pandas
import pandas as pd
```

```python
#Membaca Dataset
dataset='https://raw.githubusercontent.com/asepmuhidin/MK-Data-Mining-UPB/main/dataset/cereal.csv'
df=pd.read_csv(dataset)
```

```python
#Menampilkan 10 Baris Pertama
df.head(10)
```

```python
#MenampilkanDetail
df.tail(10)
```

```python
#Menampilkan Sample
df.sample(10)
```

```python
# Pisahkan data numerikal dan kategorikal
numerical_df = df.select_dtypes(include=['number'])
categorical_df = df.select_dtypes(exclude=['number'])
```

```python
# Tampilkan hasil pemisahan data
print("\nData Numerikal:")
print(numerical_df.head())
```

```python
#MenampilkanDataKategorikal
print("\nData Kategorikal:")
print(categorical_df.head())
```

```python
# Mencari data dengan nilai lemak tertinggi
highest_fat = df[df['fat'] == df['fat'].max()]

# Menampilkan hasil
print("Data dengan Nilai Lemak Tertinggi:")
print(highest_fat)
```

```python
# Mencari data dengan nilai lemak terendah
lowest_fat = df[df['fat'] == df['fat'].min()]

# Menampilkan hasil
print("Data dengan Nilai Lemak Terendah:")
print(lowest_fat)
```

```python
# Mencari satu baris dengan nilai lemak paling rendah
lowest_fat_min = df[df['fat'] == df['fat'].min()].nsmallest(1, 'fat')

# Menampilkan hasil
print("Data dengan Nilai Lemak Paling Rendah:")
print(lowest_fat_min)
```
